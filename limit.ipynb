```python
import math
import matplotlib.pyplot as plt
from tabulate import tabulate
```

```python
def visualize_limit(input_list, output_list):
    plt.scatter(input_list, output_list)
    plt.xlabel("Input x")
    plt.ylabel("Output f(x)")
    plt.title("Visualization of Limit")
    plt.grid(True)
    plt.show()

    rows = list(zip(input_list, output_list))
    print(tabulate(rows, headers=["x", "y"], tablefmt="github"))
```

```python
def f(x):
    return math.sin(x)
```

```python
input_list = list()
output_list = list()
num_points = 20
step_size = 0.0001

for i in range(num_points):

    input = (num_points - i) * step_size
    input_list.append(input)

    output = f(input)
    output_list.append(output)

visualize_limit(input_list, output_list)
```
